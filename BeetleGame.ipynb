```python
#Import the random module to generate random numbers
import random

#Function for beetle game dice roll
def dice_roll():
    print("\n***Welcome to the Beetle Game!!***\n")
    gameon = True
    while gameon:
        for current_player in range(players_num):
            player_input = input(f"Player {current_player+1}, press Enter to roll the dice. Press Q to quit the game.")
            
            # Quit the game
            if player_input == "Q" or player_input == "q":
                print('\nGame Aborted!!!')
                winner = None
                final_score = 0
                gameon = False
                break
                
            dice_value = random.randint(1, 6)
            players_rollcount[current_player] += 1
            print(f"You rolled a : {dice_value}")
            if players_beetles[current_player][dice_value] > 0:
                if dice_value == 1 and players_beetles[current_player][5]==1:
                    print("Sorry, you cannot draw a mouth without a head")
                elif dice_value == 2 and players_beetles[current_player][5]==1:
                    print("Sorry, you cannot draw eyes without a head")
                elif dice_value == 3 and players_beetles[current_player][5]==1:
                    print("Sorry, you cannot draw antenna without a head")
                elif dice_value == 4 and players_beetles[current_player][6]==1:
                    print("Sorry, you cannot draw legs without a body")
                else:
                    print(f"Great! You can draw the {beetle_bodyparts[dice_value]}")

                    #Reduce the required beetle part count for the current player
                    players_beetles[current_player][dice_value] -= 1
                    
            else:
                print(f"You have already drawn {beetle_bodyparts[dice_value]}")

            for body_part in players_beetles[current_player]:
                if players_beetles[current_player][body_part]:
                    print(f"You need {players_beetles[current_player][body_part]} {beetle_bodyparts[body_part]}")
            print()
                    
            #Check if player has drawn a beetle and won the game
            if sum(players_beetles[current_player].values()) == 0:
                gameon = False
                winner = current_player
                final_score = players_rollcount[current_player]
                print(f"Congratulations Player {current_player+1}, you have won the beetle game!!")
                break
    return winner, final_score


if __name__ == "__main__":
    
    beetle_bodyparts = {1 : 'Mouth', 2 : 'Eye', 3 : 'Antenna', 4 : 'Leg', 5 : 'Head', 6 : 'Body'}
    players_beetles = []
    players_rollcount = []

    players_num = int(input("Please enter number of players playing the game : "))

    #For each players create a dictionary in the list which contains track on number of beetle parts the players have.
    for i in range(players_num):
        players_beetles.append({
            1: 1,  #Mouth
            2: 2,  #Eyes
            3: 2,  #Antenna
            4: 6,  #Legs
            5: 1,  #Head
            6: 1   #Body
            })
    
        #Append 0 as starting score for each player
        players_rollcount.append(0)
    
    # Call the dice roll function
    winner_num,score = dice_roll()
    
    #Write the result in a file
    if score > 0 :
        result = f"The winner is Player {winner_num+1} who won the beetle game in {score} rolls of the dice."
        file1 = open('Beetlegame_Result.txt','w')
        file1.write(result)
        file1.close()
```
